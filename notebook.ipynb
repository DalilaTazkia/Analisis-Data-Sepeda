## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

```python
# ── 5. Cek Outlier dengan IQR ─────────────────────────────────────────────────
def detect_outliers_iqr(series, label):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_out = ((series < lower) | (series > upper)).sum()
    print(f"  {label}: Q1={Q1:.1f}, Q3={Q3:.1f}, IQR={IQR:.1f}, "
          f"lower_bound={lower:.1f}, upper_bound={upper:.1f}, outlier={n_out}")

print("=== Deteksi Outlier (IQR) – day.csv ===")
for col in ['casual', 'registered', 'cnt']:
    detect_outliers_iqr(df_day[col], col)

print()
print("=== Deteksi Outlier (IQR) – hour.csv ===")
for col in ['casual', 'registered', 'cnt']:
    detect_outliers_iqr(df_hour[col], col)

print()
print("   Masalah 2 : Terdapat outlier pada kolom 'casual', 'registered', dan 'cnt'.")
print("   Outlier ini kemungkinan merupakan hari-hari ekstrem (event besar / cuaca sangat buruk)")
print("   sehingga TIDAK dihapus, melainkan dicatat dan dianalisis lebih lanjut.")
```

### Cleaning Data

#### Fixing ... problem

```python
def clean_bike_df(df):
    df = df.copy()

    # 1. Konversi dteday ke datetime
    df['dteday'] = pd.to_datetime(df['dteday'])

    # 2. Tambahkan label yang readable untuk kolom kategorikal
    df['season_label'] = df['season'].map({
        1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
    })
    df['weather_label'] = df['weathersit'].map({
        1: 'Cerah/Berawan Tipis',
        2: 'Berkabut/Berawan',
        3: 'Hujan/Salju Ringan',
        4: 'Hujan Lebat/Es'
    })
    df['year_label'] = df['yr'].map({0: '2011', 1: '2012'})
    df['day_type'] = df.apply(
        lambda r: 'Hari Libur/Akhir Pekan' if (r['holiday'] == 1 or r['workingday'] == 0)
                  else 'Hari Kerja', axis=1
    )

    # 3. Denormalisasi suhu (nilai dalam dataset dinormalisasi ke [0,1])
    #    temp: dibagi 41 (max °C), atemp: dibagi 50, hum: dibagi 100, windspeed: dibagi 67
    df['temp_c']      = df['temp']      * 41
    df['atemp_c']     = df['atemp']     * 50
    df['hum_pct']     = df['hum']       * 100
    df['wind_kph']    = df['windspeed'] * 67

    return df

df_day_clean  = clean_bike_df(df_day)
df_hour_clean = clean_bike_df(df_hour)

print("Data berhasil dibersihkan dan diperkaya.")
print(f"Kolom baru day  : {[c for c in df_day_clean.columns if c not in df_day.columns]}")
print(f"Kolom baru hour : {[c for c in df_hour_clean.columns if c not in df_hour.columns]}")
```

## Exploratory Data Analysis (EDA)

### Explore ...

**Pertanyaan 1: Pola Peminjaman Per Jam (Hari Kerja vs. Hari Libur)**

```python
# ── Rata-rata peminjaman per jam berdasarkan tipe hari ────────────────────────
hourly_pattern = (
    df_hour_clean
    .groupby(['hr', 'day_type'])['cnt']
    .mean()
    .reset_index()
    .rename(columns={'cnt': 'avg_cnt'})
)

# Pivot untuk perbandingan mudah
hourly_pivot = hourly_pattern.pivot(index='hr', columns='day_type', values='avg_cnt')
print("=== Rata-rata peminjaman per jam ===")
print(hourly_pivot.round(1).to_string())
```

```python
# ── Jam puncak ────────────────────────────────────────────────────────────────
for day_type in hourly_pivot.columns:
    peak_hr = hourly_pivot[day_type].idxmax()
    peak_val = hourly_pivot[day_type].max()
    print(f"[{day_type}] Jam puncak: {peak_hr:02d}:00 → rata-rata {peak_val:.1f} peminjaman")
```

```python
# ── Statistik ringkasan per rentang waktu ─────────────────────────────────────
def jam_kategori(hr):
    if 6 <= hr <= 9:   return 'Pagi (06-09)'
    elif 10 <= hr <= 15: return 'Siang (10-15)'
    elif 16 <= hr <= 20: return 'Sore (16-20)'
    else:               return 'Malam/Dini Hari'

df_hour_clean['period'] = df_hour_clean['hr'].apply(jam_kategori)

period_stats = (
    df_hour_clean
    .groupby(['period', 'day_type'])['cnt']
    .agg(['mean', 'sum', 'count'])
    .round(1)
)
print("=== Statistik per Periode Waktu ===")
print(period_stats)
```

**Pertanyaan 2: Pengaruh Cuaca & Musim terhadap Peminjaman Harian**

```python
# ── Rata-rata & total peminjaman per kondisi cuaca ───────────────────────────
weather_stats = (
    df_day_clean
    .groupby('weather_label')['cnt']
    .agg(['mean', 'median', 'sum', 'count'])
    .round(1)
    .sort_values('mean', ascending=False)
)
weather_stats.columns = ['Rata-rata', 'Median', 'Total', 'Jumlah Hari']
print("=== Statistik Peminjaman per Kondisi Cuaca ===")
print(weather_stats)
```

```python
# ── Rata-rata peminjaman per musim ────────────────────────────────────────────
season_stats = (
    df_day_clean
    .groupby('season_label')['cnt']
    .agg(['mean', 'sum', 'count'])
    .round(1)
    .sort_values('mean', ascending=False)
)
season_stats.columns = ['Rata-rata', 'Total', 'Jumlah Hari']
print("=== Statistik Peminjaman per Musim ===")
print(season_stats)
```

```python
# ── Korelasi faktor lingkungan dengan cnt (harian) ───────────────────────────
corr_cols = ['temp_c', 'atemp_c', 'hum_pct', 'wind_kph', 'cnt']
corr_matrix = df_day_clean[corr_cols].corr()
print("=== Korelasi Faktor Lingkungan vs. Jumlah Peminjaman ===")
print(corr_matrix['cnt'].sort_values(ascending=False).to_string())
```

```python
# ── Cross-tab cuaca × musim ───────────────────────────────────────────────────
cross_weather_season = (
    df_day_clean
    .groupby(['season_label', 'weather_label'])['cnt']
    .mean()
    .unstack()
    .round(0)
)
print("=== Rata-rata Peminjaman (Musim × Cuaca) ===")
print(cross_weather_season.to_string())
```

**Lanjutan – Analisis Tren Bulanan & Clustering Hari**

```python
# ── Tren bulanan per tahun ────────────────────────────────────────────────────
monthly_trend = (
    df_day_clean
    .groupby(['year_label', 'mnth'])['cnt']
    .sum()
    .reset_index()
)
print("=== Total Peminjaman Bulanan per Tahun ===")
print(monthly_trend.pivot(index='mnth', columns='year_label', values='cnt').round(0))
```

```python
# ── Clustering manual hari berdasarkan jumlah peminjaman (binning) ────────────
bins   = [0, 1500, 3000, 4500, 6000, 9000]
labels = ['Sangat Rendah', 'Rendah', 'Sedang', 'Tinggi', 'Sangat Tinggi']
df_day_clean['demand_cluster'] = pd.cut(df_day_clean['cnt'], bins=bins, labels=labels)

cluster_summary = (
    df_day_clean
    .groupby('demand_cluster', observed=True)
    .agg(
        jumlah_hari   = ('cnt', 'count'),
        avg_cnt       = ('cnt', 'mean'),
        avg_temp      = ('temp_c', 'mean'),
        avg_hum       = ('hum_pct', 'mean'),
        pct_workday   = ('workingday', 'mean')
    )
    .round(2)
)
cluster_summary['pct_workday'] = (cluster_summary['pct_workday'] * 100).round(1)
print("=== Ringkasan Cluster Permintaan ===")
print(cluster_summary.to_string())
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pola Peminjaman Per Jam (Hari Kerja vs. Hari Libur)

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Pola Peminjaman Sepeda Per Jam\nHari Kerja vs. Hari Libur/Akhir Pekan',
             fontsize=15, fontweight='bold', y=1.02)

colors = {'Hari Kerja': '#2196F3', 'Hari Libur/Akhir Pekan': '#FF9800'}

# ── Subplot kiri: Line chart ───────────────────────────────────────────────────
ax1 = axes[0]
for day_type, color in colors.items():
    subset = hourly_pattern[hourly_pattern['day_type'] == day_type]
    ax1.plot(subset['hr'], subset['avg_cnt'],
             marker='o', markersize=5, linewidth=2.5,
             label=day_type, color=color)

# Tandai jam puncak
for day_type, color in colors.items():
    col_data = hourly_pivot[day_type]
    peak_hr  = col_data.idxmax()
    peak_val = col_data.max()
    ax1.annotate(f'Puncak\n{peak_hr:02d}:00\n({peak_val:.0f})',
                 xy=(peak_hr, peak_val),
                 xytext=(peak_hr + (1.5 if day_type == 'Hari Kerja' else -3), peak_val - 40),
                 fontsize=8.5, color=color,
                 arrowprops=dict(arrowstyle='->', color=color, lw=1.5))

ax1.set_xlabel('Jam', fontsize=12)
ax1.set_ylabel('Rata-rata Peminjaman', fontsize=12)
ax1.set_title('Line Chart – Pola Per Jam', fontsize=13)
ax1.set_xticks(range(0, 24))
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.4)

# Arsir jam sibuk pagi & sore untuk hari kerja
ax1.axvspan(7, 9, alpha=0.08, color='#2196F3', label='_')
ax1.axvspan(16, 19, alpha=0.08, color='#2196F3', label='_')

# ── Subplot kanan: Heatmap per jam per hari dalam minggu ──────────────────────
ax2 = axes[1]
weekday_hour = (
    df_hour_clean
    .groupby(['weekday', 'hr'])['cnt']
    .mean()
    .unstack()
)
weekday_names = ['Minggu','Senin','Selasa','Rabu','Kamis','Jumat','Sabtu']
weekday_hour.index = [weekday_names[i] for i in weekday_hour.index]

sns.heatmap(weekday_hour, ax=ax2, cmap='YlOrRd',
            fmt='.0f', linewidths=0.3,
            cbar_kws={'label': 'Rata-rata Peminjaman'})
ax2.set_title('Heatmap – Jam × Hari dalam Minggu', fontsize=13)
ax2.set_xlabel('Jam', fontsize=12)
ax2.set_ylabel('Hari', fontsize=12)
ax2.tick_params(axis='x', rotation=0, labelsize=8)

plt.tight_layout()
plt.savefig('viz1_pola_per_jam.png', bbox_inches='tight', dpi=150)
plt.show()
print(" Visualisasi 1 tersimpan.")
```

### Pertanyaan 2: Pengaruh Cuaca & Musim terhadap Peminjaman Harian

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Pengaruh Kondisi Cuaca & Musim\nterhadap Jumlah Peminjaman Sepeda Harian',
             fontsize=15, fontweight='bold', y=1.02)

# ── Subplot kiri: Barplot cuaca ────────────────────────────────────────────────
ax1 = axes[0]
weather_order = weather_stats.sort_values('Rata-rata', ascending=False).index.tolist()
palette_w = ['#4CAF50', '#FF9800', '#F44336', '#9C27B0'][:len(weather_order)]

weather_avg = (
    df_day_clean
    .groupby('weather_label')['cnt']
    .mean()
    .reindex(weather_order)
    .reset_index()
)
bars = ax1.bar(range(len(weather_avg)),
               weather_avg['cnt'],
               color=palette_w[:len(weather_avg)],
               edgecolor='white', linewidth=1.2, width=0.6)

# Tambahkan nilai di atas bar
for bar_obj, val in zip(bars, weather_avg['cnt']):
    ax1.text(bar_obj.get_x() + bar_obj.get_width()/2,
             bar_obj.get_height() + 60,
             f'{val:,.0f}', ha='center', va='bottom', fontsize=10.5, fontweight='bold')

ax1.set_xticks(range(len(weather_avg)))
ax1.set_xticklabels(weather_avg['weather_label'], fontsize=10)
ax1.set_ylabel('Rata-rata Peminjaman per Hari', fontsize=12)
ax1.set_title('Rata-rata Peminjaman per Kondisi Cuaca', fontsize=13)
ax1.set_ylim(0, weather_avg['cnt'].max() * 1.2)
ax1.grid(axis='y', alpha=0.4)

# ── Subplot kanan: Boxplot musim ──────────────────────────────────────────────
ax2 = axes[1]
season_order = ['Spring', 'Summer', 'Fall', 'Winter']
palette_s = ['#81C784', '#FFB74D', '#EF5350', '#64B5F6']
season_data = [
    df_day_clean[df_day_clean['season_label'] == s]['cnt'].values
    for s in season_order
]
bp = ax2.boxplot(season_data, patch_artist=True, notch=False,
                 medianprops=dict(color='black', linewidth=2.5))
for patch, color in zip(bp['boxes'], palette_s):
    patch.set_facecolor(color)
    patch.set_alpha(0.75)

# Overlay strip plot (jitter)
for i, (s, color) in enumerate(zip(season_order, palette_s), start=1):
    y = df_day_clean[df_day_clean['season_label'] == s]['cnt']
    x = np.random.normal(i, 0.06, size=len(y))
    ax2.scatter(x, y, alpha=0.25, s=10, color=color)

ax2.set_xticklabels(season_order, fontsize=11)
ax2.set_ylabel('Jumlah Peminjaman Harian', fontsize=12)
ax2.set_title('Distribusi Peminjaman per Musim (Boxplot)', fontsize=13)
ax2.grid(axis='y', alpha=0.4)

plt.tight_layout()
plt.savefig('viz2_cuaca_musim.png', bbox_inches='tight', dpi=150)
plt.show()
print(" Visualisasi 2 tersimpan.")
```

## Conclusion & Recommendation

**Kesimpulan 1 – Pola Peminjaman Per Jam**

Terdapat perbedaan pola yang sangat mencolok antara **hari kerja** dan **hari libur/akhir pekan**:

- **Hari Kerja** menunjukkan pola **bimodal** (dua puncak): puncak pertama pada jam **08:00** dan puncak kedua yang lebih tinggi pada jam **17:00–18:00**, mencerminkan pola **commuter** (perjalanan ke/dari tempat kerja).
- **Hari Libur/Akhir Pekan** menunjukkan pola **unimodal** (satu puncak), dengan permintaan tertinggi pada kisaran **11:00–14:00**, mencerminkan aktivitas **rekreasi** di tengah hari.
- Dari heatmap, terlihat bahwa **Rabu–Jumat pukul 08:00 dan 17:00–18:00** adalah slot waktu dengan permintaan tertinggi secara konsisten.

**Kesimpulan 2 – Pengaruh Cuaca & Musim**

- **Kondisi cuaca** berpengaruh sangat signifikan: peminjaman pada hari **Cerah/Berawan Tipis** rata-rata **~4.876 unit**, sedangkan **Hujan/Salju Ringan** hanya **~1.803 unit** (turun ≈63%). Hari dengan **Hujan Lebat/Es** hampir tidak ada peminjaman.
- **Musim Fall (Gugur)** memiliki rata-rata peminjaman **tertinggi**, diikuti Summer, Winter, dan Spring. Hal ini kemungkinan karena suhu yang ideal di musim gugur (tidak terlalu panas/dingin).
- Korelasi suhu (`temp_c`) dengan jumlah peminjaman sebesar **r ≈ +0.63**, menandakan hubungan positif yang kuat: semakin hangat, semakin banyak peminjaman.
- Kelembaban (`hum_pct`) berkorelasi **negatif (r ≈ -0.10)** dan kecepatan angin (`wind_kph`) berkorelasi negatif moderat, artinya cuaca lembab dan berangin mengurangi minat peminjaman.

**Kesimpulan 3 – Tren & Clustering**

- Total peminjaman **2012 secara konsisten lebih tinggi** dari 2011 di setiap bulan, menunjukkan **pertumbuhan yang signifikan** (~70% YoY). Puncak terjadi di bulan **Juni–September** (musim panas–gugur).
- Dari clustering, **60%+ hari** masuk kategori permintaan **Sedang–Tinggi** (3.000–6.000), namun terdapat cluster **Sangat Rendah** yang didominasi hari cuaca buruk dan musim dingin.

Dasar Analisis: Peminjaman turun >50% saat cuaca buruk.

Dasar Analisis: Distribusi musiman peminjaman.
